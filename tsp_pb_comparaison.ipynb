### Comparaison of tsp algorism performance 

```python
# loading librarys
import numpy as np
import matplotlib.pyplot as plt
import time
import kmeans_tsp_solver
import asymmetric_tsp_solver
import evolutionary_tsp_solver
```

```python
# load the data in lab2 directory
# n = [10, 20, 50, 100, 200, 500, 1000]
n = [10, 20, 50, 100, 200, 500, 1000]
restarts = [200, 100, 50, 30, 30, 10, 3]
g_pb = []
r1_pb = []
r2_pb = []

for i in n:
    data = np.load("lab2/problem_g_"+str(i)+".npy")
    g_pb.append(data)
    data = np.load("lab2/problem_r1_"+str(i)+".npy")
    r1_pb.append(data)
    data = np.load("lab2/problem_r2_"+str(i)+".npy")
    r2_pb.append(data)

print("Data loaded")
```

#### Using k-means symetric TSP solver on g problems

```python
g_sol_sym = []
g_time_sym = []
g_cost_sym = []

for i in range(len(n)):
    print(f"\n{'='*60}")
    print(f"Processing g problem with n={n[i]} cities (symmetric solver)")
    print(f"{'='*60}")
    
    start_time = time.time()
    
    final_tour, cost_breakdown = kmeans_tsp_solver.kmeans_tsp_solve(
        g_pb[i],
        max_clusters=max(n[i]//5, 10),
        restarts_per_cluster=30,
        general_restarts=restarts[i],
        auto_clusters=True,
        verbose=False
    )
    
    elapsed_time = time.time() - start_time
    
    g_sol_sym.append(final_tour)
    g_time_sym.append(elapsed_time)
    g_cost_sym.append(cost_breakdown['final'])
    
    print(f"\nCompleted in {elapsed_time:.2f}s, final cost: {cost_breakdown['final']:.2f}")

print("\n" + "="*60)
print("Summary for g problems (symmetric solver):")
print("="*60)
for i in range(len(n)):
    print(f"n={n[i]:4d}: cost={g_cost_sym[i]:10.2f}, time={g_time_sym[i]:6.2f}s")
```

#### Using asymmetric TSP solver on r1 problems

```python
r1_sol_asym = []
r1_time_asym = []
r1_cost_asym = []

for i in range(len(n)):
    print(f"\n{'='*60}")
    print(f"Processing r1 problem with n={n[i]} cities (asymmetric solver)")
    print(f"{'='*60}")
    
    start_time = time.time()
    
    final_tour, cost_breakdown = asymmetric_tsp_solver.asymmetric_tsp_solve(
        r1_pb[i],
        max_clusters=max(n[i]//5, 10),
        restarts_per_cluster=30,
        final_restarts=restarts[i],
        auto_clusters=True,
        verbose=False
    )
    
    elapsed_time = time.time() - start_time
    
    r1_sol_asym.append(final_tour)
    r1_time_asym.append(elapsed_time)
    r1_cost_asym.append(cost_breakdown['final'])
    
    print(f"\nCompleted in {elapsed_time:.2f}s, final cost: {cost_breakdown['final']:.2f}")

print("\n" + "="*60)
print("Summary for r1 problems (asymmetric solver):")
print("="*60)
for i in range(len(n)):
    print(f"n={n[i]:4d}: cost={r1_cost_asym[i]:10.2f}, time={r1_time_asym[i]:6.2f}s")
```
