### Comparaison of tsp algorism performance 

```python
# loading librarys
import numpy as np
import matplotlib.pyplot as plt
import time
import kmeans_tsp_solver
import asymmetric_tsp_solver
```

```python
# load the data in lab2 directory
# n = [10, 20, 50, 100, 200, 500, 1000]
n = [10, 20, 50]
g_pb = []
r1_pb = []
r2_pb = []

for i in n:
    data = np.load("lab2/problem_g_"+str(i)+".npy")
    g_pb.append(data)
    data = np.load("lab2/problem_r1_"+str(i)+".npy")
    r1_pb.append(data)
    data = np.load("lab2/problem_r2_"+str(i)+".npy")
    r2_pb.append(data)

print("Data loaded")
```

#### Using k-means symetric TSP solver on g problems

```python
g_sol_sym = []
g_time_sym = []
g_cost_sym = []

for i in range(len(n)):
    print(f"\n{'='*60}")
    print(f"Processing g problem with n={n[i]} cities (symmetric solver)")
    print(f"{'='*60}")
    
    start_time = time.time()
    
    final_tour, cost_breakdown = kmeans_tsp_solver.kmeans_tsp_solve(
        g_pb[i],
        max_clusters=max(n[i]//5, 10),
        restarts_per_cluster=30,
        auto_clusters=True,
        verbose=True
    )
    
    elapsed_time = time.time() - start_time
    
    g_sol_sym.append(final_tour)
    g_time_sym.append(elapsed_time)
    g_cost_sym.append(cost_breakdown['final'])
    
    print(f"\nCompleted in {elapsed_time:.2f}s, final cost: {cost_breakdown['final']:.2f}")

print("\n" + "="*60)
print("Summary for g problems (symmetric solver):")
print("="*60)
for i in range(len(n)):
    print(f"n={n[i]:4d}: cost={g_cost_sym[i]:10.2f}, time={g_time_sym[i]:6.2f}s")
```

#### Using asymmetric TSP solver on r1 problems

```python
r1_sol_asym = []
r1_time_asym = []
r1_cost_asym = []

for i in range(len(n)):
    print(f"\n{'='*60}")
    print(f"Processing r1 problem with n={n[i]} cities (asymmetric solver)")
    print(f"{'='*60}")
    
    start_time = time.time()
    
    final_tour, cost_breakdown = asymmetric_tsp_solver.asymmetric_tsp_solve(
        r1_pb[i],
        max_clusters=max(n[i]//5, 10),
        restarts_per_cluster=30,
        final_restarts=10,
        auto_clusters=True,
        verbose=True
    )
    
    elapsed_time = time.time() - start_time
    
    r1_sol_asym.append(final_tour)
    r1_time_asym.append(elapsed_time)
    r1_cost_asym.append(cost_breakdown['final'])
    
    print(f"\nCompleted in {elapsed_time:.2f}s, final cost: {cost_breakdown['final']:.2f}")

print("\n" + "="*60)
print("Summary for r1 problems (asymmetric solver):")
print("="*60)
for i in range(len(n)):
    print(f"n={n[i]:4d}: cost={r1_cost_asym[i]:10.2f}, time={r1_time_asym[i]:6.2f}s")
```

#### Using asymmetric TSP solver on r2 problems

```python
r2_sol_asym = []
r2_time_asym = []
r2_cost_asym = []

for i in range(len(n)):
    print(f"\n{'='*60}")
    print(f"Processing r2 problem with n={n[i]} cities (asymmetric solver)")
    print(f"{'='*60}")
    
    start_time = time.time()
    
    final_tour, cost_breakdown = asymmetric_tsp_solver.asymmetric_tsp_solve(
        r2_pb[i],
        max_clusters=max(n[i]//5, 10),
        restarts_per_cluster=30,
        final_restarts=10,
        auto_clusters=True,
        verbose=True
    )
    
    elapsed_time = time.time() - start_time
    
    r2_sol_asym.append(final_tour)
    r2_time_asym.append(elapsed_time)
    r2_cost_asym.append(cost_breakdown['final'])
    
    print(f"\nCompleted in {elapsed_time:.2f}s, final cost: {cost_breakdown['final']:.2f}")

print("\n" + "="*60)
print("Summary for r2 problems (asymmetric solver):")
print("="*60)
for i in range(len(n)):
    print(f"n={n[i]:4d}: cost={r2_cost_asym[i]:10.2f}, time={r2_time_asym[i]:6.2f}s")
```

#### Visualization of Results

```python
# Plot execution times
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Time comparison
ax1.plot(n, g_time_sym, 'o-', label='g (symmetric)', linewidth=2, markersize=8)
ax1.plot(n, r1_time_asym, 's-', label='r1 (asymmetric)', linewidth=2, markersize=8)
ax1.plot(n, r2_time_asym, '^-', label='r2 (asymmetric)', linewidth=2, markersize=8)
ax1.set_xlabel('Number of cities (n)', fontsize=12)
ax1.set_ylabel('Execution time (seconds)', fontsize=12)
ax1.set_title('Execution Time Comparison', fontsize=14, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_xscale('log')
ax1.set_yscale('log')

# Cost comparison
ax2.plot(n, g_cost_sym, 'o-', label='g (symmetric)', linewidth=2, markersize=8)
ax2.plot(n, r1_cost_asym, 's-', label='r1 (asymmetric)', linewidth=2, markersize=8)
ax2.plot(n, r2_cost_asym, '^-', label='r2 (asymmetric)', linewidth=2, markersize=8)
ax2.set_xlabel('Number of cities (n)', fontsize=12)
ax2.set_ylabel('Tour cost', fontsize=12)
ax2.set_title('Solution Cost Comparison', fontsize=14, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_xscale('log')
ax2.set_yscale('log')

plt.tight_layout()
plt.show()

# Print detailed comparison table
print("\n" + "="*80)
print("DETAILED COMPARISON TABLE")
print("="*80)
print(f"{'n':>6} | {'g (sym)':>25} | {'r1 (asym)':>25} | {'r2 (asym)':>25}")
print(f"{'':>6} | {'Cost':>12} {'Time':>12} | {'Cost':>12} {'Time':>12} | {'Cost':>12} {'Time':>12}")
print("-"*80)
for i in range(len(n)):
    print(f"{n[i]:6d} | {g_cost_sym[i]:12.2f} {g_time_sym[i]:11.2f}s | "
          f"{r1_cost_asym[i]:12.2f} {r1_time_asym[i]:11.2f}s | "
          f"{r2_cost_asym[i]:12.2f} {r2_time_asym[i]:11.2f}s")
print("="*80)
```
